# Numpyによる正規乱数の生成

```python
from numpy.random import *
from matplotlib import pyplot as plt
import seaborn as sns
```

## 1次元正規乱数

```python
# 平均10, 標準偏差5の正規乱数を10万個生成
mu = 10
sigma = 5
num = 100000

values = normal(mu, sigma, num)
```

```python
# ヒストグラム
plt.hist(values, bins=50)
plt.show()
```

## 2次元正規乱数

```python
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
num = 10000

values = multivariate_normal(mu, sigma, num)
```

```python
# 散布図
sns.jointplot(values[:,0], values[:,1])
plt.show()
```
